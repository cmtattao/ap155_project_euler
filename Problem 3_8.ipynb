```python
# Problem 3.8: This problem studies the second derivative of the function: f(x) = (1-cos(x))/(x^2) which you encountered in problem 2.8. We take x = 0.0004.

import numpy as np
import matplotlib.pyplot as plt

x = 0.004

def f_func(x):
    y = (1-np.cos(x))/(x**2)
    return y

f_prime = (x*np.sin(x) + 2*np.cos(x) - 2)/(x**3)

f_sd = ((x**2)*np.cos(x) - 6*np.cos(x) - (4*x)*np.sin(x) + 6)/(x**4)

def forward_diff(f_func, x, h):
    fd = (f_func(x+h) - f_func(x))/(h)
    return fd

def central_diff(func, x, h):
    cd = (f_func(x+h) - 2*f_func(x) + f_func(x-h))/(h**2)
    return cd
```

```python
# Part a) Start by analytically evaluating the second derivative, and plugging in x = 0.004.

# f'(x) = (x*sin(x) + 2*cox(x) - 2)/(x^3)
# f''(x) = ((x^2 - 6)cos(x) - 4x*sin(x) + 6)/(x^4)

hs = [10**(-i) for i in range(1,7)]
hs = np.logspace(-7, -1, 200)
ans = [abs(central_diff(f_func, x, h) - f_sd) for h in hs]

print("Analytic Second Derivative", f_sd)

for i in range(len(hs)):
    print (ans[i])
```

```python
# Part b) Given h the values 10e-1, 10e-2, ..., 10e-6 and produce a log-log plot of the absolute error in the first central difference approximation to the second derivative.
fig, ax = plt.subplots()
ax.loglog(hs, ans)
ax.set_title("Central Difference (Absolute Error)")
plt.savefig("Problem 3-8.png")
```
