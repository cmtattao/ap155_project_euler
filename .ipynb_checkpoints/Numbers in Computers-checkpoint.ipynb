```python
import matplotlib.pyplot as plt
import numpy as np

from decimal import Decimal
from fractions import Fraction

# Classic floating point quirks example
print(0.1 + 0.2)

# How about increasing precision?
print(np.float128(0.1) + np.float128(0.2))

# How bad is decreasing precision?
print(np.float16(0.1) + np.float16(0.2))

# Closest we could get to fixed-point arithmetic: decimal
print(Decimal(0.1) + Decimal(0.2))

# How about using an explicit Fraction object type?
print(float(Fraction(1, 10) + Fraction(1, 5)))
```

```python
# Problem 2.6 version 2

a = np.float16(1)
c = np.float16(1)


for i in [4,5,6,7,8,9]:
    b = np.float128(10**i)
    
    x_plus = (-b + np.sqrt(b**2 - 4*a*c))/(2*a)
    x_minus = (-b - np.sqrt(b**2 - 4*a*c))/(2*a)
    
    x_plus_alt = c/(a*x_minus)

    print(f"b = 10^{i}")
   # print(x_plus, x_plus_alt)
    print("Log-error: ", np.log10(np.abs(x_plus - x_plus_alt)))
```

```python
fig, ax = plt.subplots()
ax.set_xscale("log")
ax.set_yscale("log")

for dtype in [np.float64, np.float128]:
    a = dtype(1)
    c = dtype(1)

    b = np.logspace(2, 9, 100, dtype=dtype)

    x_plus = (-b + np.sqrt(b**2 - 4*a*c))/(2*a)
    x_minus = (-b - np.sqrt(b**2 - 4*a*c))/(2*a)
    x_plus_alt = c/(a*x_minus)

    ax.scatter(b, np.abs(x_plus - x_plus_alt), label=str(dtype))

ax.legend()
ax.grid(True)
```

```python
from math import*
def f(x):     #these are all the same function using different identities  
    a = (1-(sin(x)/tan(x)))/(x**2)
    b = (1-(sin(2*x)/(2*sin(x))))/(x**2)
    c = (1-((1-((tan(x/2))**2))/(1+(tan(x/2))**2)))/(x**2)
    d = (sin(x)**2+cos(x)**2-cos(x))/(x**2)
    e = (sin(x)**2+cos(x)**2-(sin(2*x)/(2*sin(x))))/(x**2)
    g = (1 - cos(x))/(x**2) 
    return a, b, c, d, e, g

print(f(1.2e8))
```
