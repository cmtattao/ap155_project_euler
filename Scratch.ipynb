```python
import numpy as np
import matplotlib.pyplot as plt

# Number of points and grid setup
n = 128
x = np.linspace(0, 2*np.pi, n, endpoint=False)  # grid points
h = x[1] - x[0]  # grid spacing

# Define functions
f = 2 + 5*np.sin(x) + 0.1*np.sin(30*x)   # noisy function
g = 2 + 5*np.sin(x)                      # smooth function

# Analytical derivatives
f_prime = 5*np.cos(x) + 3*np.cos(30*x)   # derivative of f
g_prime = 5*np.cos(x)                    # derivative of g

# (b) Forward difference derivative (adjacent points)
fd_adj = (f[1:] - f[:-1]) / h
x_fd_adj = x[:-1]  # derivative at left endpoint

# (c) Forward difference with "twice-removed" points
fd_twice = (f[2:] - f[:-2]) / (2*h)
x_fd_twice = x[:-2]

# ----------------- PLOTTING -----------------

# (a) Plot f(x) vs g(x)
plt.figure(figsize=(10,4))
plt.plot(x, f, label="f(x) = 2 + 5 sin(x) + 0.1 sin(30x)")
plt.plot(x, g, '--', label="g(x) = 2 + 5 sin(x)")
plt.xlabel("x")
plt.ylabel("Function value")
plt.title("(a) f(x) and g(x)")
plt.legend()
plt.grid(True)
plt.show()

# (b) and (c) Derivatives
plt.figure(figsize=(10,5))
plt.plot(x, f_prime, label="Analytical f'(x)")
plt.plot(x, g_prime, '--', label="Analytical g'(x)")
plt.plot(x_fd_adj, fd_adj, 'o', markersize=3, label="Forward diff (adjacent)")
plt.plot(x_fd_twice, fd_twice, 'x', markersize=3, label="Forward diff (twice-removed)")
plt.xlabel("x")
plt.ylabel("Derivative value")
plt.title("(b) and (c) Derivative comparisons")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problem 3.25: This problem examines the derivative of a noisy function
# f(x) = 2 + 5 sin(x) + 0.1 sin(30x)
# compared with g(x) = 2 + 5 sin(x) (no noise)

import numpy as np
import matplotlib.pyplot as plt

# Define functions
def f_func(x):
    return 2 + 5*np.sin(x) + 0.1*np.sin(30*x)

def g_func(x):
    return 2 + 5*np.sin(x)

# Analytical derivatives
def f_prime(x):
    return 5*np.cos(x) + 3*np.cos(30*x)

def g_prime(x):
    return 5*np.cos(x)

# Forward-difference approximation
def forward_diff(f, x, h):
    return (f(x+h) - f(x)) / h

# Grid setup
N = 128
x = np.linspace(0, 2*np.pi, N, endpoint=False)
h = x[1] - x[0]

# Function values
fx = f_func(x)
gx = g_func(x)

# Numerical derivatives
forward_diff_h = (np.roll(fx, -1) - fx) / h              # step h
forward_diff_2h = (np.roll(fx, -2) - fx) / (2*h)         # step 2h

# --- Plot (a): f(x) vs g(x)
plt.figure(figsize=(10, 3))
plt.plot(x, fx, label="f(x) with noise")
plt.plot(x, gx, label="g(x) smooth", linestyle="--")
plt.legend()
plt.title("Comparison of f(x) and g(x)")
plt.xlabel("x")
plt.ylabel("Function value")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Plot (b) and (c): derivatives
plt.figure(figsize=(10, 4))
plt.plot(x, f_prime(x), label="Analytical f'(x)", linewidth=1)
plt.plot(x, g_prime(x), label="Analytical g'(x)", linestyle="--", linewidth=2)
plt.plot(x, forward_diff_h, label="Forward diff f'(x), h", alpha=0.7)
plt.plot(x, forward_diff_2h, label="Forward diff f'(x), 2h", alpha=0.7)

plt.legend()
plt.title("Derivative Comparison")
plt.xlabel("x")
plt.ylabel("Derivative value")
plt.grid(True)
plt.tight_layout()
plt.show()
```
