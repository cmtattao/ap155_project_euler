```python
import numpy as np
import matplotlib.pyplot as plt

def f_func(x):
    y = (1 - np.cos(x))/(x**2)

    return y
    
def forward_diff(func, x, h):
    y_prime = (x*np.sin(x) + 2*np.cos(x) - 2)/x**3

    return y_prime

```

```python
h = 10**-2
x = np.arange(0.004, 10, h)

y = f_func(x)

y_xp1 = np.roll(y, -1)
y_xm1 = np.roll(y, +1)


y_fd = (y_xp1 + y_xm1 - 2*y)/(h**2)

plt.plot(x[1:-1], y_fd[1:-1])
```

```python

h = 0.1
x_min = 0.004
x_max = 10

x = np.arange(x_min, x_max, h)
y = (1 - np.cos(x))/(x**2)

fx_p1 = np.roll(y, -1)
fx_m1 = np.roll(y, +1)

deriv = (fx_p1 - fx_m1)/(2*h)
```

```python
plt.plot(x, y, label="orig")
plt.plot(x[1:-1], deriv[1:-1], label="deriv")

plt.legend()
```
