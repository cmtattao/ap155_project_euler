```python
t = np.arange(0, 10, 0.05)
freq = 1.0

y = np.sin(2*np.pi*freq*t)

fig, ax = plt.subplots()

shifted_y = np.sin(2*np.pi*freq*t - 2.0)
ax.plot(t, y, label="not shifted")
ax.plot(t, shifted_y,  label="is shifted")

ax.legend()
```

```python
#Basic Numpy/Matplotlib Exercise 1

t = np.arange(0, 1, 0.001)
freq1 = 1.0
freq2 = 10.0

y_1 = np.sin(2*np.pi*freq1*t)
y_2 = np.sin(2*np.pi*freq2*t)

fig, ax = plt.subplots()
ax.plot(t, y_1*y_2, label="Product of two signals")

ax.legend()
```

```python
t = np.linspace(0, 1, 100)
y = np.sqrt(1 - t**2)

fig, ax = plt.subplots()

ax.plot(t, y)

for i in range(100):
    x_i = np.random.random()
    y_i = np.random.random()

    if x_i**2 + y_i**2 < 1:
        ax.scatter(x_i, y_i)
```

```python
t = np.linspace(0, 1, 100)
y = np.sqrt(1 - t**2)

fig, ax = plt.subplots()

ax.plot(t, y)

c_in = 0
c_out = 0

for i in range(10000):
    x_i = np.random.random()
    y_i = np.random.random()

    if x_i**2 + y_i**2 < 1:
        ax.scatter(x_i, y_i)
        c_in = c_in + 1
    else:
        c_out = c_out + 1

print(4*(c_in/(c_in+c_out)))
        
```

```python

N = 1000000
x_i = np.random.random(size = N)
y_i = np.random.random(size = N)

is_inside = x_i**2 + y_i**2 < 1

fig, ax = plt.subplots()

ax.scatter(x_i[is_inside], y_i[is_inside])

print(4*np.sum(is_inside)/N)
```
